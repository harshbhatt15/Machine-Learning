```python
import pandas as pd 
from matplotlib import pyplot as plt
%matplotlib inline
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
df = pd.read_csv(r'D:\Machine Learning\Supervised Machine Learning\classification\Logistic Algorithm\Diabetes\diabetes.csv')
df
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['Age'], df['Glucose'], marker='o')
plt.xlabel("Age")
plt.ylabel("Glucose")
plt.title("Features(Logistic Regression)")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X = df[['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']]
Y = df.Outcome
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y,test_size=0.4,random_state=35)
```

```python
X_test
```

```python
Y_train
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
```

```python
model.fit(X_train, Y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,Y_test)
```

```python
#probabilities betwween 0 and 1
model.predict_proba(X_test)
```

```python
y_pred = model.predict_proba(X_test)[::,1]

fpr, tpr, _ = roc_curve(Y_test,  y_pred)
```

```python
plt.plot(fpr, tpr, color='blue', label='ROC Curve')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')  # diagonal line
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```

```python
auc = roc_auc_score(Y_test, y_pred)
print("AUC:", auc)
```
