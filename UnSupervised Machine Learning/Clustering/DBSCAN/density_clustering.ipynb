```python
X_scaled = df[['Weight', 'Height']].values
```

```python
plt.scatter(X_scaled[:,0], X_scaled[:,1])
plt.xlabel('Scaled Weight')
plt.ylabel('Scaled Height')
plt.title('Scaled Weight vs Height')
plt.show()
```

```python
db = DBSCAN(eps=0.5, min_samples=5)

db.fit(X_scaled)
```

```python
labels = db.labels_
```
